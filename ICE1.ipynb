```python
image = cv2.imread('NicolasCage.jpg')
```

```python
plt.imshow(image)
```

#Smurf

```python
## convert to grayscale
bnw = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(bnw, cmap = 'gray')
```

## successfully converted

```python
cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = cascade.detectMultiScale(bnw, 1.5, 5)
faces
```

```python
(x,y,w,h) = faces[0]
face_image = cv2.rectangle(image, (x,y), (x+w, y+h),(200,0,0),2)
plt.imshow(face_image)
```
